```python
# Create the count plot
ax = sns.countplot(data=df, x='NumberOfFollowups', hue='ProdTaken')

# Calculate total counts for each NumberOfFollowups
total_counts = df['NumberOfFollowups'].value_counts().sort_index()
counts = df.groupby(['NumberOfFollowups', 'ProdTaken']).size().unstack(fill_value=0)

# Add percentage labels
for container in ax.containers:
    labels = []
    for i, bar in enumerate(container):
        height = bar.get_height()
        number_followups = df['NumberOfFollowups'].unique()[i // 2]
        percentage = (height / total_counts[number_followups]) * 100
        labels.append(f'{percentage:.2f}%')
    ax.bar_label(container, labels=labels)

plt.show()
```
